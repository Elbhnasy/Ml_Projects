```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
file="Detailes.csv"
f=open(file,"w",encoding="utf-8")
#creating file for saving data in it 
```

```python
Header="JopTitle,             CompanyName     ,Skills \n"
f.write(Header)                      #columns title 
for page in range (1,12):
    url=f"https://wuzzuf.net/search/jobs/?a=hpb&q=python&start={page}"
    html=urlopen(url)
    soup=BeautifulSoup(html,"html.parser")
    container=soup.find_all("div",{"class":"css-1gatmva e1v1l3u10"})
    for i in container:
        JopTitle=i.find("h2",{"class":"css-m604qf"}).text.strip()
        CompanyName=i.find("a",{"class":"css-17s97q8"}).text.strip()
        Skills=i.find("div",{"class":"css-y4udm8"}).text.strip()
        
        f.write(f"{JopTitle},{CompanyName},  {Skills}\n")
f.close()        
```
