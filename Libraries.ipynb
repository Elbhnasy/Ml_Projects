The multivariate normal, multinormal or Gaussian distribution is a generalization of the one-dimensional normal distribution to higher dimensions. Such a distribution is specified by its mean and covariance matrix. These parameters are analogous to the mean (average or “center”) and variance (standard deviation, or “width,” squared) of the one-dimensional normal distribution.

```python
import scipy as sc
```

```python
def equection(x):
    return x**2+10*np.sin(x)
```

 *fminbound* Bounded minimization for scalar functions.

```python
#scipy.optimize.fminbound(func, x1, x2, args=(), xtol=1e-05, maxfun=500, full_output=0, disp=1)
# func: Objective function to be minimize
#x1, x2 :The optimization bounds.
#xtol The convergence tolerance
'''''disp 0 : no message printing. 1 : non-convergence notification messages only.
2 : print a message on convergence too.
3 : print iteration results.'''
min_bound=sc.optimize.fminbound(equection,x1=-20,x2=20,xtol=.5,
                                full_output=True,disp=3)
```

```python
from scipy.optimize import curve_fit
```

```python
def func(x, a, b, c):
    return a * np.exp(-b * x) + c
```

curve_fit: Use non-linear least squares to fit a function, f, to data.

```python
xdata = np.linspace(0, 4, 50)
y = func(xdata, 2.5, 1.3, 0.5)
y_noise = 0.2 * np.random.normal(size=xdata.size)
ydata = y + y_noise
plt.plot(xdata, ydata, 'b-', label='data')
popt, pcov = curve_fit(func, xdata, ydata)
plt.plot(xdata, func(xdata, *popt), 'r-',
         label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))
popt, pcov = curve_fit(func, xdata, ydata, bounds=(0, [3., 1., 0.5]))
plt.plot(xdata, func(xdata, *popt), 'g--',
         label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
